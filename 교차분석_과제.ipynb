# 교차분석 과제 
## titanic데이터에서 sex와 class 간의 동질성 검정을 진행하세요. 

```python
import pandas as pd 
```

```python
data = pd.read_csv('titanic.csv')
data
```

1. 귀무가설 : Sex 의 분포는 Class 와 관계없이 동일하다
2. 대립가설 : Sex 의 분포는 Class 에 관계없이 동일하지 않다

```python
table = pd.crosstab(index = data["sex"], columns = data["class"])
```

```python
from scipy.stats import chi2_contingency

chi, p, df, expect = chi2_contingency(table)

print("chi 값",chi)
print("p 값",p)
print("df 값",df)
print("expect 값",expect)
```

p value 값이 0.05보다 한참 작으므로 Sex의 분포는 Class에 관계없이 동일하지 않음 즉 귀무가설 채택
